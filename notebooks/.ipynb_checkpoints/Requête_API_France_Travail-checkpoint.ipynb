```python
import requests
%time

# Paramètres de l'API
API_URL = "https://api.francetravail.io/partenaire/offresdemploi/v2/offres/search"
ACCESS_TOKEN = access_token  # Obtenez-le via OAuth
all_results = []  # Liste pour stocker toutes les offres

for codeROME in codes_ROME:
    # Configuration des paramètres de pagination
    page = 1
    size = 50  # Nombre d'offres par page (maximum permis par l'API)
    print(f"*********  Code ROME n°: {codeROME} **********")
    while True:
        # Préparer la requête
#         params = {"appellation" : appellation, "page": page, "size": size}
        params = {"codeROME" : codeROME, "page": page, "size": size}
        headers = {"Authorization": f"Bearer {ACCESS_TOKEN}",
                  'Accept': "application/json"}

        # Envoyer la requête
        response = requests.get(API_URL, headers=headers, params=params)

        # Vérifier le succès de la requête
        if response.status_code != 200:
            print(f"Erreur : {response.status_code}, {response.text}")
            break

        # Extraire les données JSON
        data = response.json()
        offres = data.get("resultats", [])  # La clé exacte dépend de la structure JSON de l'API

        # Si aucune offre n'est retournée, c'est la fin
        if not offres:
            print("Toutes les pages ont été parcourues.")
            break

        # Ajouter les résultats à la liste
        all_results.extend(offres)
        print(f"Page {page} récupérée : {len(offres)} offres.")

        # Passer à la page suivante
        page += 1

    # Résumé
    print(f"Nombre total d'offres récupérées : {len(all_results)}")
```
